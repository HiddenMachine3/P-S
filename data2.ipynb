```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataSet = pd.read_excel("./Corr_Reg_dataset.xlsx", sheet_name="data2")
```

```python
temp_X = np.array(dataSet["Temperature (X)"])
yield_Y = np.array(dataSet["% Yield (Y)"])
```

```python
plt.scatter(temp_X, yield_Y)
```

```python
x_mean = np.mean(temp_X)
y_mean = np.mean(yield_Y)
```

```python
slope = np.sum((temp_X - x_mean) * (yield_Y - y_mean)) / np.sum((temp_X - x_mean) ** 2)
intercept = y_mean - slope * x_mean
```

```python
pred = slope * temp_X + intercept
```

```python
plt.scatter(temp_X, yield_Y)
plt.plot(temp_X, pred, color="red")
plt.ylabel("% Yield (Y)")
plt.xlabel("Temperature (X)")
plt.grid(True)
plt.title(f"Least Squares Regression: Slope={slope:.3f} Intercept={intercept:.3f}")
```
